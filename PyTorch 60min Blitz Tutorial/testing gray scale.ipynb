```python
import torch
import torchvision

import numpy as np
import matplotlib.pyplot as plt

from PIL import Image


# opens original as PIL image
pil_panda = Image.open('panda.jpg').resize( (224,224) )

# converts original to numpy array
np_panda = np.array(pil_panda)

# converts original to tensor
panda_tensor = torch.from_numpy(np_panda)
```

```python
# transform PIL image to gray scale
grey_panda_1 = torchvision.transforms.functional.to_grayscale(pil_panda, num_output_channels=1)

# converts grayscale PIL to numpy array
np_grey_panda_1 = np.array(grey_panda_1)

plt.imshow(grey_panda)
```

```python
# manually make original into gray panda

r_chan = panda_tensor[:,:,0].numpy()
g_chan = panda_tensor[:,:,1].numpy()
b_chan = panda_tensor[:,:,2].numpy()

result = (0.3 * r_chan + 0.59 * g_chan + 0.11 * b_chan)/3.0

plt.imshow(result)

#pil_result = torchvision.transforms.ToPILImage()(result)

pil_result = Image.fromarray(result.astype('uint8'))
pil_result.save('tensor_gray_test.JPG')
```
